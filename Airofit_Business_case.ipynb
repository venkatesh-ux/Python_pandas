```python
label1 = [0,18,40,50] # 0-18 (teen), 18-40 (Adult), 4-50 (Middle age)
groups = ['Teen','Adult','Middle age']
```

```python
df['Age_bins'] = pd.cut(df['Age'],label1,labels=groups)
df
df['Age_bins'].value_counts()
```

```python
sns.countplot(df,x='Age_bins',hue='Product')
plt.show()
```

```python
df['Income'].describe()
```

```python
labels2 = [20000,40000,70000,110000]
groups2 = ['Low','Medium','high']
```

```python
df['Income_bin'] = pd.cut(df['Income'],labels2,labels=groups2)
df.head()
```

```python
sns.countplot(df,x='Income_bin',hue='Product')
plt.show()
```
